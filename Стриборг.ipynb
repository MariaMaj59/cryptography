В тексте ГОСТа название данной операции звучит как сложение в кольце вычетов по модулю 2 в степени n.
Два исходных 64-байтовых вектора представляются как два больших числа, далее они складываются, и переполнение, если оно появляется, отбрасывается:

```python
# Нелинейное биективное преобразование (преобразование S)

def S(text):
  res = 0
  for j in range(63,-1,-1):
      bit8 = text // 2**(8*j) % (2**8)
      res += table[bit8] << (j*8)
  return res
```

```python
#  Преобразование P (перестановка)

def P(text):
  res = 0
  for j in range(63,-1,-1):
      bit8 = text // 2**(8*j) % (2**8)
      res += bit8 << (Tau[j]*8)
  return res
```

```python
#   Линейное преобразование множества двоичных векторов (Преобразование L)

def L(text):
  res = 0
  for i in range(7,-1,-1):
    bit8 = text // 2**(64*i) % (2**64)
    for j in range(63, -1, -1):
      if((bit8 >> j) & 1):
        res ^= A[63-j] << (i*64)
  return res
```

```python
# Функция сжатия

def E(K, text):
  res = K ^ text
  for i in range(12):
    res = L(P(S(res)))
    K = L(P(S(K ^ C[i]))) # вычисление ключей
    res ^= K
  return res


def g(N, h, text):
  return E(L(P(S(h ^ N))), text) ^ h ^ text
```

```python
def verification(text):
  if(len(text) < 512):
    text = '1' + text
  return int('0b' + text, base = 2)
```

```python
def vec256():
  h = ''
  for i in range(64):
    h += '00000001'
  return int('0b' + h, base = 2)
```

```python
def hesh512(text, inter):
  tlen = len(text)
  text = verification(text)
  global h, N
  h = g(N, h, text)
  N = (N + tlen) % 2**(8*64)
  inter = (inter + text) % 2**(8*64)
  return inter
```

```python

def stribog (h_size, text):
  global h, N
  # Первый этап
  if (h_size == 256):
    h = vec256()
  elif(h_size == 512):
    h = 0

  # Второй этап
  inter = 0
  while(len(text) > 0):
    inter = hesh512(text[-512:], inter)
    text = text[:-512]
  
  # Третий этап
  h = g(0, h, N)  
  h = g(0, h, inter) >> (512 - h_size)
  return h
```
